```python
x_data = df['Category']
y_dat = df['Size']

plt.scatter(x_data,y_dat)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Relation ship between size and category')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data =df ,x = "ship-state")
plt.xlabel('ship-state')
plt.ylabel('Count')
plt.title('Distribution state')
plt.xticks(rotation = 90)
plt.show()
```

```python
# top 10 state status 
top_10_state = df['ship-state'].value_counts().head(10)
# plot count city by state 
plt.figure(figsize=(12,6))
sns.countplot(data =df[df['ship-state'].isin(top_10_state.index)] ,x = "ship-state")
plt.xlabel('ship-state')
plt.ylabel('Count')
plt.title('Distribution state')
plt.xticks(rotation = 90)
plt.show()
```
